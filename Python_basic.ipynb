1.  What is Python, and why is it popular?
- Python is a high-level, interpreted programming language known for its readability, simplicity, and vast library support.
Popularity reasons:

 - Easy to learn & read (English-like syntax)

 - Large standard library and third-party modules

 - Cross-platform compatibility

 - Versatile — used in web dev, data science, automation, AI, etc.

2.  What is an interpreter in Python?
-  An interpreter is a program that reads and executes Python code line by line, converting it into machine-readable instructions.

3.  What are pre-defined keywords in Python?
-  Keywords are reserved words with special meaning in Python (e.g., if, for, while, True). They cannot be used as identifiers (variable names).

4.  Can keywords be used as variable names?
-  No, keywords are reserved and cannot be used as variable names.

5.  What is mutability in Python?
-  Mutability means the ability of an object’s content to be changed after creation.

6.  Why are lists mutable, but tuples are immutable?

-  List: Designed for dynamic data, allowing modifications.

-  Tuple: Used for fixed data storage, ensuring data integrity.

7.  Difference between == and is operators in Python:

-  == → Checks value equality (whether values are the same).

-  is → Checks identity (whether both refer to the same object in memory).

8.  What are logical operators in Python?

-  and → True if both operands are True.

-  or → True if at least one operand is True.

-  not → Inverts the boolean value.

9.   What is type casting in Python?
-  Converting a variable from one data type to another.

10.  Difference between implicit and explicit type casting:

-  Implicit: Done automatically by Python (e.g., int to float).

-  Explicit: Manually done using functions like int(), float().

11.  Purpose of conditional statements in Python:
-  To control program flow based on certain conditions.

12.  How does the elif statement work?
-   Checks another condition if the previous if/elif condition is false.

13.  Difference between for and while loops:

-  for loop: Iterates over a sequence (known range).

-  while loop: Runs until a condition becomes false.

14.  Scenario where while is more suitable:
-   When the number of iterations is unknown beforehand — e.g., reading user input until they type "exit".

```python
# 1. Hello, World!
print("Hello, World!")

# 2. Display name and age
name = "John"
age = 25
print(f"My name is {name} and I am {age} years old.")

# 3. Print all Python keywords
import keyword
print(keyword.kwlist)

# 4. Check if a word is a Python keyword
word = input("Enter a word: ")
print(f"{word} is a keyword" if keyword.iskeyword(word) else f"{word} is not a keyword")

# 5. List vs Tuple mutability
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_list[0] = 99  # Works
# my_tuple[0] = 99  # Error

# 6. Mutable vs Immutable arguments
def modify_data(mutable, immutable):
    mutable.append(4)
    immutable += (4,)
    print("Inside function:", mutable, immutable)

lst = [1, 2, 3]
tpl = (1, 2, 3)
modify_data(lst, tpl)
print("Outside function:", lst, tpl)

# 7. Arithmetic operations
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sum:", a+b)
print("Difference:", a-b)
print("Product:", a*b)
print("Quotient:", a/b)

# 8. Logical operators
x, y = True, False
print(x and y, x or y, not x)

# 9. String to int, float, boolean
user_input = input("Enter a value: ")
try:
    print(int(user_input), float(user_input), bool(user_input))
except ValueError:
    print("Invalid input for integer conversion. Please enter a numerical value.")

# 10. Type casting list elements
lst = ["1", "2", "3"]
lst = [int(i) for i in lst]
print(lst)

# 11. Positive, negative, or zero
num = int(input("Enter a number: "))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")

# 12. For loop 1 to 10
for i in range(1, 11):
    print(i)

# 13. Sum of even numbers between 1 and 50
total = sum(i for i in range(1, 51) if i % 2 == 0)
print("Sum:", total)

# 14. Reverse string using while loop
s = input("Enter a string: ")
rev = ""
i = len(s) - 1
while i >= 0:
    rev += s[i]
    i -= 1
print("Reversed:", rev)

# 15. Factorial using while loop
n = int(input("Enter a number: "))
fact = 1
while n > 0:
    fact *= n
    n -= 1
print("Factorial:", fact)
```
